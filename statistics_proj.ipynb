# Statistical Analysis and Visualization with Python


Objective:
This project aims to perform statistical analysis and data visualization using Python. You will generate a synthetic 
dataset, perform descriptive and inferential statistics, and visualize the results using Matplotlib and Seaborn.


```python
#Importing Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#defining number of sample
sample_no=1000
```

```python
#Generating Synthetic Data
Age = np.random.normal(loc=35, scale=10, size=sample_no)
Height = np.random.normal(loc=170, scale=15, size=sample_no)
Weight = np.random.normal(loc=70, scale=10, size=sample_no)
Gender = np.random.choice(['Male','Female'], size=sample_no, p=[0.5,0.5])
Income = np.random.normal(loc=50000, scale=15000, size=sample_no)

#Combining in a dataframe
data=pd.DataFrame({
    'Age' : Age,
    'Height' : Height,
    'Weight' : Weight,
    'Gender' : Gender,
    'Income' : Income
})

print(data.head())
```

```python
#To save the dataset

data.to_csv('Synthetic_ds.csv', index=False)
print('Dataset saved')
```

# Pearson Correlation Coefficient


```python
#Pearson Correlation Coefficient


columns = ['Age','Weight','Height','Income']
selected_data = data[columns]
pearson= selected_data.corr(method='pearson')
print(pearson)
```

```python
#HeatMap
plt.figure(figsize=(10,8))
sns.heatmap(pearson, annot=True, cmap='Spectral')
plt.title('Correlation')
plt.show()
```

- There is no significant relation btw any variable.

# Inferential Statistics

```python
#split the data

Male = data[data['Gender']=='Male']
Female = data[data['Gender']=='Female']
```

```python
# T-Test
from scipy.stats import ttest_ind

result = ttest_ind(Male['Income'], Female['Income'], equal_var=False, nan_policy='omit')

print(f'{'Income'} : t-statistics = {result.statistic:.4f}, p_value = {result.pvalue:.4f}')
```

- As per the t-statistics value - it shows mean value of male income is less than mean value of female.
- as p value- as it is > than 0.05. so the differnce in income is not statistically significant.
